```python
## Let's create a new column called number of total missed appointments for this patient
frequency = pd.value_counts(df.PatientId.values, sort=True, normalize=True)
## First this gets us the frequency of visits to the facilit(ies)
apptfreqdict={}
for entry in frequency.index:
    apptfreqdict[entry] = frequency[entry]

apptfreqdict
##Now let's turn this into a dictionary
##2.724571e+11    0.000009
##9.462664e+14    0.000027
##2.386225e+14    0.000018
##2.523155e+14    0.000009
##6.626949e+14    0.000036
```

```python
### This creates a new column called freq of visits at the very end of the ddataframe
df['ApptFreq'] = pd.Series(df.PatientId, index=df.index)

#df.FreqOfVisits.dtype

##df['FreqOfVisits'].map(freqdict) 
##df.sample(n=20)
```

```python
#now this changes the columns at the end from patient ids to the normalized frequency of visits
df['ApptFreq']= df['ApptFreq'].map(apptfreqdict)
```
